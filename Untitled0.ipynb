```python
df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salário': 'salário',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residência',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}, inplace=True)

display(df.columns)
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})

display(df['senioridade'].value_counts())
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})

display(df['contrato'].value_counts())
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})

display(df['tamanho_empresa'].value_counts())
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})

display(df['remoto'].value_counts())
```

```python
import numpy as np

df_salarios = pd.DataFrame({
    'nome': ["Ana", "bruno", "carlos", "daniele" , "val"],
    'salario': [4000, np.nan , 5000, np.nan , 100000]
})
#subistitui a media e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''calcula a mediana e substitui
'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())


df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "dia" : ['segunda','terça','quarta', 'quinta', 'sexta'],
    "temperatura" : [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["temperatura"].ffill()

df_temperaturas
```

```python
import plotly.express as px
```

```python
df_avg_salary_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()

fig = px.bar(df_avg_salary_senioridade,
             x='senioridade',
             y='usd',
             title='Média Salarial por Senioridade',
             labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual (USD)'})

fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray': ordem})

fig.show()
```

```python
remoto_contagem = df_limpo ['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']
fig = px.pie(remoto_contagem,
             names ='tipo_trabalho',
             values = 'quantidade',
             title = 'Proporção dos tipos de trabalho')

fig.show()
```

```python
df_avg_salary_country = df_limpo.groupby('empresa')['usd'].mean().reset_index()
df_avg_salary_country = df_avg_salary_country.sort_values(by='usd', ascending=False)

# Crie um dicionário mapeando códigos de país para cores
cores_paises = {
    'US': 'blue',   # Exemplo: Azul para os EUA
    'BR': 'green', # Exemplo: Verde para o Brasil
    'CA': 'red',   # Exemplo: Vermelho para o Canadá
    'GB': 'purple' # Exemplo: Roxo para o Reino Unido
    # Adicione mais países e cores aqui conforme necessário
}

fig = px.bar(df_avg_salary_country,
             x='empresa',
             y='usd',
             title='Média Salarial por País',
             labels={'empresa': 'País', 'usd': 'Média Salarial Anual (USD)'},
             color='empresa', # Use a coluna 'empresa' para determinar a cor
             color_discrete_map=cores_paises # Use o dicionário de cores personalizadas
            )

fig.show()
```
